```python
import numpy as np
steps = np.arange(0,2001,250)
loss = np.array([4.2657, 1.9678, 1.7214, 1.5935, 1.5193, 1.4905, 1.4698, 1.4638, 1.4777])
ppl = np.exp(loss)
steps.size == loss.size
```

```python
import matplotlib.pyplot as plt


#plot perplexity on step
plt.figure(figsize=(8, 5), dpi=300)
plt.scatter(steps, ppl, s=70, color="k")

plt.yscale("log")
plt.xlabel("Step", fontsize=15)
plt.ylabel("Perplexity", fontsize=15)
plt.title("Perplexity vs Step", fontsize=17)
plt.gca().tick_params(axis='both', which='both', labelsize=15)

for x, y in zip(steps, ppl): plt.annotate(f"{y:.2f}", (x, y), xytext=(0, 6), textcoords="offset points", fontsize=10, ha="left", va="bottom")

plt.ylim(3,100)
plt.tight_layout()


plt.show()
```
